Train and Test Data

```python
x=data.drop("species",axis=1)
y=data["species"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
m=LogisticRegression()
```

```python
m.fit(x_train,y_train)
```

```python
x_pred=m.predict(x_train)
accuracy=accuracy_score(y_train,x_pred)
print('Accuracy score of training data is:',accuracy)
```

```python
x_pred=m.predict(x_test)
test_accuracy=accuracy_score(y_test,x_pred)
print('Accuracy score of testing data is:',test_accuracy)
```

```python
y_pred=m.predict(x_test)
matrix=confusion_matrix(y_test,y_pred)
matrix
```
